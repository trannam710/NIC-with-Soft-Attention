```python
from caption import *

img_name = '000000316807'
img_path = 'UIT-ViIC/test/' + img_name + '.jpg'
model = './BEST_syllable_pretrain_checkpoint_uitviic_5_cap_per_img_5_min_word_freq.pth.tar'
word_map = './caption data syllable/word_map_syllable.json'

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
checkpoint = torch.load(model, map_location=str('cpu'))
decoder = checkpoint['decoder']
decoder = decoder.to(device)
decoder.eval()
encoder = checkpoint['encoder']
encoder = encoder.to(device)
encoder.eval()

# Load word map (word2ix)
with open(word_map, 'r') as j:
    word_map = json.load(j)
rev_word_map = {v: k for k, v in word_map.items()}  # ix2word

# Encode, decode with attention and beam search
seq, alphas = caption_image_beam_search(encoder, decoder, img_path, word_map, 5)
alphas = torch.FloatTensor(alphas)

image = Image.open(img_path)

words = [rev_word_map[ind] for ind in seq]
    
cap  =' '.join(words[1:-1])
plt.text(0, 1, '%s' % (cap), color='black', backgroundcolor='white', fontsize=12)
plt.imshow(image)
plt.show()
```
